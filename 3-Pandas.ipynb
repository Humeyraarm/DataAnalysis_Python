# Creating a Pandas Series

pd.Series(data=None, index=None, dtype=None, name=None, copy=False, fastpath=False)

One-dimensional ndarray with axis labels (including time series).

```python
import numpy as np
import pandas as pd
```

```python
labels = ['a','b','c']
my_list = [10, 20, 30]
arr = np.array([10, 20, 30])
d = {'a': 10, 'b': 20, 'c': 30}
```

**Using Lists**

```python
pd.Series(labels)
```

```python
pd.Series(my_list)
```

```python
pd.Series([10, 20, 30])
```

**Using arrays**

```python
arr
```

```python
pd.Series(data=arr,index = labels)
```

**Using Dictionaries**

```python
d                                #dictionary has already indexes 
```

```python
pd.Series(d,index=['q','c','y'])  # if there is same index uses.if not return nan.
```

**Using Scalar Value**

```python
pd.Series(data = "humeyra")
```

```python
pd.Series(data = "humeyra",index = range(10,13))
```

```python
pd.Series(data = 10, index= ['a', 'b', 'c'])
```

# **Indexing Pandas Series**

```python
ser2 = pd.Series([1,2,5,4],index = ['USA', 'Germany','Italy', 'Japan'])
```

```python
ser1.sort_index()
```

```python
ser2.sort_values()
```

```python
ser1[1]
```

```python
ser1['Germany']
```

```python
ser1.Germany    #for string data(without blank) only
```

```python
ser1+ser2       #add values of same indexes
```
