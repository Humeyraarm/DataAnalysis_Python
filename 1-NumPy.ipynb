# **NumPy**
NumPy (or Numpy) is a Linear Algebra Library for Python, the reason it is so important for Data Science with Python is that almost all of the libraries in the PyData Ecosystem rely on NumPy as one of their main building blocks.

Numpy is also incredibly fast, as it has bindings to C libraries.

A data attribute is exactly as it sounds; it's data, it gives simply a property. A method is a procedure, an action.

```python
import numpy as np
```

**Numpy Arrays**

NumPy arrays are the main way we will use Numpy throughout the course. Numpy arrays essentially come in two flavors: vectors and matrices. Vectors are strictly 1-d arrays and matrices are 2-d (but you should note a matrix can still have only one row or one column).

```python
my_list = [1,2,3]
my_list
```

```python
my_array = np.array(my_list)
my_array
```

```python
np.array([1,2,3])
```

```python
my_list = [[1,2,3,4],[1,2,3,4],[2,3,4,5]]
my_list
```

**Random**

Numpy also has lots of ways to create random number arrays:

**rand**

Create an array of the given shape and populate it with random samples from a uniform distribution over [0, 1).

(0, 5) = 1, 2, 3, 4

(0, 5] = 1, 2, 3, 4, 5

[0, 5) = 0, 1, 2, 3, 4

[0, 5] = 0, 1, 2, 3, 4, 5

```python
np.random.rand(2)   #always btw 0 and 1
```

```python
np.random.rand(5,5)
```

```python
plt.hist(np.random.rand(555555), bins= 40)    # uniform dist 
plt.show();

# first 55, then 555, 5555, 55555... the more sample the mor
```

**randn**

Return a sample (or samples) from the "standard normal" distribution. Unlike rand which is uniform:

```python
np.random.randn(2)    #not only btw 0 and 1. depends on mean=0 std=1
```

```python
np.random.randn(5,5)
```

```python
plt.hist(np.random.randn(5555), bins=55)
plt.show()
```

```python
np.random.randn(5555).mean()  # Standart normal distrubition gives mean of 0
```

```python
np.random.randn(5555).std()**2  #Standart normal distrubition gives variance of 1.
```

# **Splitting of the Arrays**

np.split(array, indices_or_sections, axis=0)

```python
x = np.array([1, 2, 3, 99, 99, 3, 2, 1])
x
```

```python
np.split(x,[3,5])
```

If indices_or_sections is a 1-D array of sorted integers, the entries indicate 
where along axis the array is split. For example, [3, 5] would, for axis=0, result in

ary[:3]

ary[3:5]

ary[5:]

```python
a,b,c = np.split(x, [3,5])
```

```python
a
```

```python
b
```

```python
c
```

```python
np.split(x,[4])  # according to 4. indices
```

```python
np.split(x,4)    # divides 4 if it can be.bölünebilecekse 4 e böler.
```

The number of the elements is divisible by 4, otherwise raises an error. If the argument (here is 4) is not enclosed by square brackets, then it means dividing into 4 pieces.

```python
np.split(x,[0,5])
```
