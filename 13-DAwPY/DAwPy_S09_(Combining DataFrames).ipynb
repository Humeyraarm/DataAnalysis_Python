**``DataFrame.merge(right, how='inner', on=None, left_on=None, right_on=None, left_index=False, right_index=False, sort=False, suffixes=('_x', '_y'), copy=True, indicator=False, validate=None)``**

The [**merge()**](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.merge.html) function allows you to merge DataFrames together using a similar logic as merging SQL Tables together. For example [Source 01](https://www.brainstobytes.com/hands-on-pandas-9-merging-dataframes/), [Source 02](https://www.shanelynn.ie/merge-join-dataframes-python-pandas-index-1/), [Source 03](https://towardsdatascience.com/combining-dataframes-using-pandas-b9e2e83b9869):



```python
left = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})
   
right = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                      'C': ['C0', 'C1', 'C2', 'C3'],
                      'D': ['D0', 'D1', 'D2', 'D3']})    
```

```python
display(left, right)
```

**Let's dive into merge()**

```python
pd.merge(left, right, how='inner', on='key')

# Here, merge() function used intersection of keys (which must be found in both DataFrames) from both DataFrames, 
# similar to a SQL inner join,
# and preserved the order of the left keys.
```

```python
left2 = pd.DataFrame({'key': ['K0', 'K1', 'K4', 'K5'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})
   
right2 = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                       'C': ['C0', 'C1', 'C2', 'C3'],
                       'D': ['D0', 'D1', 'D2', 'D3']}) 
```

```python
display(left2, right2)
```

```python
pd.merge(left2, right2, how='inner', on='key')

# Two DataFrames, left2 and right2, were combined together using intersection of keys, "key" columns, from both DataFrames
# Only corresponding values from "key" common column of two DataFrame were included in the combined DataFrame 
```

```python
pd.merge(left2, right2, how='outer', on='key')

# Two DataFrames, left2 and right2, were combined together using union of keys, "key" columns, from both DataFrames
# All corresponding values from "key" common column of two DataFrames were included in the combined DataFrame 
```

```python
pd.merge(left2, right2, how='left', on='key')

# Two DataFrames, left2 and right2, were combined together using only keys, "key" column, from left DataFrame
# All corresponding values from "key" common column of "left2" DataFrame were included in the combined DataFrame 
```

```python
pd.merge(left2, right2, how='right', on='key')

# Two DataFrames, left2 and right2, were combined together using only keys,"key" column, from right DataFrame
# All corresponding values from "key" common column of "right2" DataFrame were included in the combined DataFrame 
```

**Let us do a more complicated example:**

```python
pd.merge(left3, right3, how='outer', on='key')

# Two DataFrames, left3 and right3, were combined together using union of keys, "key" columns, from both DataFrames
# All corresponding values from "key" common columns of two DataFrames were included in the combined DataFrame 

# Notice that some values were duplicated on the columns of "C" and "D"
```

```python
pd.merge(left3, right3, how='left', on='key')

# Two DataFrames, left3 and right3, were combined together using only keys,"key" column, from left DataFrame
# All corresponding values from "key" common column of "left2" DataFrame were included in the combined DataFrame

# Notice that some values were duplicated on the columns of "C" and "D"
```

```python
pd.merge(left3, right3, how='right', on='key')

# Two DataFrames, left3 and right3, were combined together using only keys, "key" column, from right DataFrame
# All corresponding values from "key" common column of "right3" DataFrame were included in the combined DataFrame

# Notice that some values were duplicated on the columns of "C" and "D"
```

As seen, similar to join, merge also combines all the columns from the two tables, with the common columns renamed with the defined suffixes.

[Pandas Merge DataFrames on Multiple Columns](https://sparkbyexamples.com/pandas/pandas-merge-two-dataframes-on-multiple-columns/)

However, merge provides three ways of flexible control over row-wise alignment.

- **The first way** is to use “on = COLUMN NAME”, here the given column must be the common column in both tables. 

- **The second way** is to use “left_on = COLUMN NAME and right_on = COLUMN NAME” , and it allows to align the two tables using two different columns.

- **The third way** is to use “left_index = True and right_index = True”, and the two tables are aligned based on their indexes.

Let's dive into more complicated examples:

```python
left4 = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                     'key2': ['K0', 'K1', 'K0', 'K1'],
                        'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3']})
    
right4 = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                       'key2': ['K0', 'K0', 'K0', 'K0'],
                       'C': ['C0', 'C1', 'C2', 'C3'],
                       'D': ['D0', 'D1', 'D2', 'D3']})
```

You can also explicitly specify the column names you wanted to use for combining. To use column names define **``"on"``** parameter of the **``merge()``** method. This also takes a list of names when you wanted to merge on multiple columns.

```python
pd.merge(left4, right4, how='inner', on=['key1', 'key2'])

# Two DataFrames, left4 and right4, were combined together using intersection of keys, "key1" and "key2" columns, from both DataFrames
# Only corresponding values from "key1" and "key2" common columns of both DataFrames were included in the combined DataFrame 

# Notice that some values were duplicated on the columns of "A" and "B" for "K1 K0" match 
```

```python
pd.merge(left4, right4, how='outer', on=['key1', 'key2'])

# Two DataFrames, left4 and right4, were combined together using union of keys, "key1" and "key2" columns, from both DataFrames
# All corresponding values from "key1" and "key2" columns of two DataFrames were included in the combined DataFrame 

# Notice that some values were duplicated on the columns of "A" and "B" for "K1 K0" match 
```

```python
pd.merge(left4, right4, how='right', on=['key1', 'key2'])

# Two DataFrames, left4 and right4, were combined together using only keys, "key1" and "key2" columns, from right DataFrame
# All corresponding values from "key1" and "key2" common columns of "right4" DataFrame were included in the combined DataFrame

# Notice that some values were duplicated on the columns of "A" and "B" for "K1 K0" match 
```

```python
pd.merge(left4, right4, how='left', on=['key1', 'key2'])

# Two DataFrames, left4 and right4, were combined together using only keys, "key1" and "key2" columns, from left DataFrame
# All corresponding values from "key1" and "key2" common columns of "left4" DataFrame were included in the combined DataFrame

# Notice that some values were duplicated on the columns of "A" and "B" for "K1 K0" match 
```

When you have column names on left and right are different and want to use these as a join column, use **``left_on``** and **``right_on``** parameters. This also takes a list of column names as values to merge on multiple columns. The left_on will be set to the name of the column in the left DataFrame and right_on will be set to the name of the column in the right DataFrame.

```python
pd.merge(left5, right5, left_on="lkey", right_on="rkey", how="inner")

# Two DataFrames, left5 and right5, were combined together using intersection of uncommon keys but including same content, "lkey" and "rkey" columns, from both DataFrames
# Only corresponding values from "lkey" and "rkey" common columns of both DataFrames were included in the combined DataFrame 
```

```python
pd.merge(left5, right5, left_on="lkey", right_on="rkey", how="outer")

# Two DataFrames, left5 and right5, were combined together using union of uncommon keys but including same content, "lkey" and "rkey" columns, from both DataFrames
# All corresponding values from "lkey" and "rkey" uncommon columns of both DataFrames were included in the combined DataFrame 
```

```python
pd.merge(left5, right5, left_on="lkey", right_on="rkey", how="left")

# Two DataFrames, left5 and right5, were combined together using uncommon key columns, but including the same content, "lkey" and "rkey" columns, from both DataFrames
# While all the values from "lkey" key column of "left5" DataFrame were included in the combined DataFrame,
# Only corresponding values from "rkey" key column of "righ5" DataFrame were included in the combined DataFrame 
```

```python
pd.merge(left5, right5, left_on="lkey", right_on="rkey", how="right")

# Two DataFrames, left5 and right5, were combined together using uncommon key columns, but including the same content, "lkey" and "rkey" columns, from both DataFrames
# While all the values from "rkey" key column of "right5" DataFrame were included in the combined DataFrame,
# Only corresponding values from "lkey" key column of "left5" DataFrame were included in the combined DataFrame 
```

⚠️ [**Warning**](https://pandas.pydata.org/pandas-docs/dev/user_guide/merging.html) ⚠️

Joining / merging on duplicate keys can cause a returned frame that is the multiplication of the row dimensions, which may result in memory overflow. It is the user’ s responsibility to manage duplicate values in keys before joining large DataFrames.

However, since df1 and/or df2 contain only one value for some observations, we do not want these values to be duplicated in the merged dataframe.

Users can use the **``validate``** argument to automatically check whether there are unexpected duplicates in their merge keys. Key uniqueness is checked before merge operations and so should protect against memory overflows. Checking key uniqueness is also a good way to ensure user data structures are as expected.

**Which to use and when to use?**

- The **``join()``** method works best when we are joining DataFrames on their indexes.
- The **``merge()``** method is more versatile and allows us to specify columns, besides the index to join on, for both DataFrames.
- We cannot use **``concat()``** if our DataFrames’ dimensions do not match along the axis in which we are trying to concatenate.
- `The **``concat()``** has inner (default) and outer joins only, whereas **``merge()``** has left, right, outer, and inner (default) joins.

## Difference between Merge, join, and concatenate

The merge() and join() methods are the DataFrame method, not a series method. The concat() method is the pandas’ method which provides the functionality to combine the pandas’ objects such as DataFrame and Series.

**Merge** 

 - The merge() function used to merge the DataFrames with database-style join such as inner join, outer join, left join, right join.
 - Combining exactly two DataFrames.
 - The join is done on columns or indexes.
 - If joining columns on columns, the DataFrame indexes will be ignored.
 - If joining indexes on indexes or indexes on a column, the index will be passed on.
 

**Join** 

 - The join() function used to join two or more pandas DataFrames/Series horizontally.
 - Join() uses merge internally for the index-on-index (by default) and column(s)-on-index join.
 - Aligns the calling DataFrame’s column(s) or index with the other objects’ index (and not the columns).
 - Defaults to left join with options for right, inner and outer join
 

**Concat** 

 - concatenate two or more pandas DataFrames/Series vertically or horizontally.
 - Aligns only on the index by specifying the axis parameter.
 - Defaults to outer join with the option for inner join
