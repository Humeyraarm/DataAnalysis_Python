**Bracket Indexing and Selection**

The simplest way to pick one or some elements of an array looks very similar to python lists:

```python
arr[8]
```

```python
# Negative indexing works like in python lists
arr[-1]
```

```python
# Get values in a range
arr[1:5]
```

```python
# arr[start:stop:step], odds
arr[1::2]
```

```python
# arr[start:stop:step], evens
arr[0::2]
```

# Indexing a 2D array (matrices)

The general format is arr_2d[row][col] or arr_2d[row,col]. I recommend usually using the comma notation for clarity.

```python
arr_2d=np.array([[5,10,15],[20,25,30],[35,40,45]])

arr_2d
```

```python
arr_2d[1]       #Indexing row
```

```python
# Format is arr_2d[row][col] or arr_2d[row,col]
# Getting individual element value

arr_2d[1][2]
```

```python
arr_2d[2,1]
```

```python
arr_2d
```

```python
# 2D array slicing
#[row,col]

arr_2d[0:2,1:]
```

```python
arr_2d[2]
```

```python
arr_2d[2,:]
```

```python
arr_2d[:,1]
```

```python
# convert 25 to 99

arr_2d[1,1]=99
arr_2d
```

```python
# default dtype is int.look at the assigning of the float value.Not work

arr_2d[1][1]=7.7
arr_2d
```

**Fancy Indexing**

Fancy indexing allows you to select entire rows or columns out of order.

```python
v=np.arange(0,30,3)
v
```

```python
v[1]
```

```python
v[7]
```

```python
# we can select separate elements using their indices in a list
index_list=[1,3,7]
v[index_list]
```

```python
v[[1,3,7]]
```

```python
# Set up matrix
arr2d=np.zeros((10,10),dtype=int)
arr2d
```

```python
length_arr = arr2d.shape[0]
length_arr
```

```python
arr2d[0]
```

```python
#Set up array
for i in range(length_arr):
    arr2d[i]=i
    
arr2d
```

```python
# we can select separate rows by indices in a list
arr2d[[2,4,6,8]]
```

# **Statistical Calculations**

np.mean(arr,axis=0) | Returns mean along specific axis

arr.sum() | Returns sum of arr

arr.min() | Returns minimum value of arr

arr.max(axis=0) | Returns maximum value of specific axis

np.var(arr) | Returns the variance of array

np.std(arr,axis=1) | Returns the standard deviation of specific axis

np.corrcoef(arr) | Returns correlation coefficient of array

```python
v=np.array([1,1,2,2,3,3,3])
v
```

```python
v.mean()
```

```python
np.median(v)
```

```python
v.std()
```

```python
v.var()
```
